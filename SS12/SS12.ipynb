```python
import pandas as pd

df = pd.read_csv('./ads_data.csv')

## lấy hệ số tương quan từ dữ liệu: lệnh corr
df["Sales Revenue (USD)"].corr(df["Advertising Cost (USD)"]) ## tương quan hoàn hảo vì < 1
```

```python
# nhúng lớp mô hình: hồi quy tuyến tính
from sklearn.linear_model import LinearRegression

# Tạo model từ class LinearRegression
model = LinearRegression()

# lấy dữ liệu X và y từ DataFrame (sử dụng iloc vì két quả trả về phải là một DataFrame)
X = df.iloc[:, 0:1]
y = df.iloc[:, 1:2]

# lệnh này huấn luyện mô hình
model.fit(X, y)

## Dự đoán: sử dụng model đã huấn luyện để dự đoán nếu X=1000 => y = ???
model.predict(pd.DataFrame([700]))


## Lấy ra độ dốc a
a = model.coef_

## lấy ra hệ số chặn b
b = model.intercept_

print(a, b) # có cả x truyền vào
```

```python
import numpy as np
import matplotlib.pyplot as plt
## Vẽ biểu đồ scatter plot (phân tán): minh họa mối quan hệ giữa chi phí quảng cáo và doanh thu bán hàng
df.plot.scatter(
    x="Advertising Cost (USD)", # chi phí quảng cáo (USD)
    y="Sales Revenue (USD)"   # Doanh thu bán hàng (USD)
)
## => Mỗi điểm trên đồ thị là một cặp giá trị (chi phí qc, doanh thu) => 35 điểm

# Tạo các giá trị X cho đường thẳng: Tạo ra 100 giá trị X nằm đều từ giá trị nhỏ nhất đến lớn nhất của trục X (vẽ đường thẳng đẹp hơn)
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1) 
# Tính toán các giá trị Y trên đường thẳng (y = ax + b)
y_line = model.intercept_ + model.coef_[0]*X_line

# Vẽ biểu đồ đường màu đỏ trên đồ thị
plt.plot(X_line, y_line, color="red", label='Best fit line')
plt.show()
```
